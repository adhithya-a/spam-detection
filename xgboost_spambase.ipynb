```python
df.rename(columns={57:'is_spam'}, inplace=True)
```

```python
y=df['is_spam']
x=df.drop(['is_spam'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
```

From Table 1: Optimal combination of parameters for the proposed XGBoost-based Spam Detector

Eta: 0.4
gamma: 0.2
maximum depth: 24
column sample: 0.75
Number of rounds (early stopping after 10): 200

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(n_estimators=200,
                  eta = 0.4,
                  gamma = 0.2,
                  max_depth = 24,
                  colsample_bytree = 0.75,
                #   eval_metric='error'
                 )
xgb.fit(x_train,y_train, eval_metric='auc', verbose=True, eval_set=[(x_test, y_test)], early_stopping_rounds=10)
```

```python
pred_xgb=xgb.predict(x_test)
score_xgb=accuracy_score(y_test,pred_xgb)
print("Accuracy: %0.2f" % score_xgb )
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test,pred_xgb)
cf_matrix
```

```python
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=30)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
classes = ['Spam', 'Ham']
plot_confusion_matrix(cf_matrix, classes)
```
